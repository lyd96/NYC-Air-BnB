```python
import pandas as pd
nyc = pd.read_csv("dataset_airbnb.csv")
print(nyc)
```

Cleaning data process:

```python
nyc['id'] = nyc['id'].astype('str')
```

```python
nyc['host_id'] = nyc['host_id'].astype('str')
```

```python
nyc['last_review'] = pd.to_datetime(nyc['last_review'], format = '%m/%d/%Y')
```

```python
nyc['host_id'].unique()
```

```python
nyc = nyc.dropna(axis=0)
```

Handle missing data

Gunakan ~ untuk menghapus data kosong atau Null/NaN9

```python
nyc[~nyc['host_name'].isna()]
```

```python
nyc = nyc[~nyc['host_name'].isna()]
nyc.info()
```

```python
nyc['price'].unique()
```

```python
nyc['price'].describe()
```

Maximal = Q3+1.5X(Q3-Q1)
Minimal = Q1-1.5X(Q3-Q1)

```python
q1 = nyc['price'].quantile(0.25)
q3 = nyc['price'].quantile(0.75)
upper_limit = q3 + 1.5 * (q3-q1)
lower_limit = q1 - 1.5 * (q3-q1)
```

```python
nyc = nyc[nyc['price']> lower_limit]
nyc = nyc[nyc['price']< upper_limit]
nyc['price'].describe()
```

```python
nyc.boxplot(column='price')
```

```python
nyc.info()
```

```python
nyc['latitude'] = nyc['latitude'].astype('str')
nyc['longitude'] = nyc['longitude'].astype('str')
```

```python
nyc['Lat, Long'] = nyc['latitude']+","+nyc['longitude']
nyc
```
